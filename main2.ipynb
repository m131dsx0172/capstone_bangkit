```python
import requests

def get_distance_with_traffic(origin, destination, api_key):
    url = "https://maps.googleapis.com/maps/api/distancematrix/json"
    params = {
        "origins": origin,
        "destinations": destination,
        "mode": "driving",
        "departure_time": "now",
        "key": api_key
    }
    response = requests.get(url, params=params)
    data = response.json()
    
    # Mendapatkan jarak dalam meter
    distance = data['rows'][0]['elements'][0]['distance']['value']
    
    # Mendapatkan waktu perjalanan menggunakan mobil dalam detik
    driving_time = data['rows'][0]['elements'][0]['duration']['value']
    
    params['traffic_model'] = 'pessimistic'
    response = requests.get(url, params=params)
    data = response.json()
    
    # Mendapatkan waktu perjalanan menggunakan mobil dengan kepadatan lalu lintas dalam detik
    driving_time_traffic = data['rows'][0]['elements'][0]['duration_in_traffic']['value']
    
    params['mode'] = 'motorcycle'
    response = requests.get(url, params=params)
    data = response.json()
    
    # Mendapatkan waktu perjalanan menggunakan motor dalam detik
    motorcycle_time = data['rows'][0]['elements'][0]['duration']['value']
    
    params['traffic_model'] = 'best_guess'
    response = requests.get(url, params=params)
    data = response.json()
    
    # Mendapatkan waktu perjalanan menggunakan motor dengan kepadatan lalu lintas dalam detik
    motorcycle_time_traffic = data['rows'][0]['elements'][0]['duration_in_traffic']['value']
    
    params['mode'] = 'walking'
    response = requests.get(url, params=params)
    data = response.json()
    
    # Mendapatkan waktu perjalanan berjalan kaki dalam detik
    walking_time = data['rows'][0]['elements'][0]['duration']['value']
    
    url = "https://maps.googleapis.com/maps/api/directions/json"
    params = {
        "origin": origin,
        "destination": destination,
        "mode": "transit",
        "key": api_key
    }
    response = requests.get(url, params=params)
    data = response.json()
    
    return distance, driving_time, driving_time_traffic, motorcycle_time, motorcycle_time_traffic, walking_time


# Contoh penggunaan
origin = "Jakarta, Indonesia"
destination = "Bandung, Indonesia"
api_key = "AIzaSyCF1UagoeTuqiJgWkNDNrUhyLy9snt59DY"

distance, driving_time, driving_time_traffic, motorcycle_time, motorcycle_time_traffic, walking_time = get_distance_with_traffic(origin, destination, api_key)

print("Jarak: ", distance, " meter")
print("Waktu perjalanan menggunakan mobil: ", driving_time, " detik")
print("Waktu perjalanan menggunakan mobil dengan lalu lintas: ", driving_time_traffic, " detik")
print("Waktu perjalanan menggunakan motor: ", motorcycle_time, " detik")
print("Waktu perjalanan menggunakan motor dengan lalu lintas: ", motorcycle_time_traffic, " detik")
print("Waktu perjalanan berjalan kaki: ", walking_time, " detik")
```

```python
import requests

def check_public_transit_availability(origin, destination, api_key):
    url = "https://maps.googleapis.com/maps/api/directions/json"
    params = {
        "origin": origin,
        "destination": destination,
        "mode": "transit",
        "key": api_key
    }
    response = requests.get(url, params=params)
    data = response.json()
    ketersediaan = 0
    if "routes" in data:
        if len(data["routes"]) > 0:
            ketersediaan = 1
        else:
            ketersediaan = 0
    else:
        print("Gagal memperoleh data rute.")
    
    return ketersediaan


# Contoh penggunaan
origin = "Jakarta, Indonesia"
destination = "Bandung, Indonesia"
api_key = "AIzaSyCF1UagoeTuqiJgWkNDNrUhyLy9snt59DY"

check_public_transit_availability(origin, destination, api_key)
```
