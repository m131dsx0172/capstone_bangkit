```python
import requests

def get_distance_with_traffic(origin, destination, api_key):
    url = "https://maps.googleapis.com/maps/api/distancematrix/json"
    params = {
        "origins": origin,
        "destinations": destination,
        "key": api_key,
        "departure_time": "now",
        "traffic_model": "best_guess"  # Anda juga dapat menggunakan "pessimistic"
    }

    response = requests.get(url, params=params)
    data = response.json()

    # Memeriksa apakah permintaan berhasil
    if response.status_code == 200 and data["status"] == "OK":
        distance = data["rows"][0]["elements"][0]["distance"]["text"]
        duration = data["rows"][0]["elements"][0]["duration"]["text"]
        traffic_duration = data["rows"][0]["elements"][0]["duration_in_traffic"]["text"]
        print("Jarak:", distance)
        print("Waktu Tempuh (Tanpa Lalu Lintas):", duration)
        print("Waktu Tempuh (Dengan Lalu Lintas):", traffic_duration)
    else:
        print("Gagal mendapatkan informasi jarak dengan lalu lintas.")
        if "error_message" in data:
            print("Pesan Error:", data["error_message"])
        elif "status" in data:
            print("Status:", data["status"])

# Menentukan tempat asal, tujuan, dan kunci API
origin = "Jakarta, Indonesia"
destination = "Bandung, Indonesia"
api_key = "AIzaSyCF1UagoeTuqiJgWkNDNrUhyLy9snt59DY"

# Memanggil fungsi untuk mendapatkan informasi jarak dengan lalu lintas
get_distance_with_traffic(origin, destination, api_key)
```

```python
list_jarak_dan_waktu_tempuh = []
def get_jarak_dan_waktu_tempuh(origin, destination, api_key):
    url = "https://maps.googleapis.com/maps/api/distancematrix/json"
    params = {
        "origins": origin,
        "destinations": destination,
        "key": api_key,
        "departure_time": "now",
        "traffic_model": "best_guess"  # Anda juga dapat menggunakan "pessimistic"
    }

    response = requests.get(url, params=params)
    data = response.json()
    ddt = []
    # Memeriksa apakah permintaan berhasil
    if response.status_code == 200 and data["status"] == "OK":
        distance = data["rows"][0]["elements"][0]["distance"]["text"]
        duration = data["rows"][0]["elements"][0]["duration"]["text"]
        traffic_duration = data["rows"][0]["elements"][0]["duration_in_traffic"]["text"]
        ddt.append(distance)
        ddt.append(duration)
        ddt.append(traffic_duration)
        list_jarak_dan_waktu_tempuh.append(ddt)
    else:
        print("Gagal mendapatkan informasi jarak dengan lalu lintas.")
        if "error_message" in data:
            print("Pesan Error:", data["error_message"])
        elif "status" in data:
            print("Status:", data["status"])
```

```python
alamat_sekolah = "SMAN 1 Sungai Penuh"
for i in range(len(df['Dusun'])):
    alamat_siswa = df['Dusun'][i] + ", Sungai Penuh"
    get_jarak_dan_waktu_tempuh(alamat_siswa, alamat_sekolah, api_key)
print(list_jarak_dan_waktu_tempuh[0:10])
```

```python
import pandas as pd

def nested_list_to_dataframe(nested_list, column_names):
    df = pd.DataFrame(nested_list, columns=column_names)
    return df

# Nama kolom
column_names = ['Distance', 'Duration', 'Traffic Duration']

# Memanggil fungsi untuk mengubah nested list menjadi DataFrame
df2 = nested_list_to_dataframe(list_jarak_dan_waktu_tempuh, column_names)

# Menampilkan DataFrame
print(df.head(10))
```

```python
# Menggabungkan df1 dan df2 secara horizontal (sumbu kolom)
concatenated_df = pd.concat([df, df2], axis=1)

print(concatenated_df.head(10))
# Menyimpan DataFrame ke file CSV
concatenated_df.to_csv('data_akhir.csv', index=False)
```
