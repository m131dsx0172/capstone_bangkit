```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
x_train = pd.read_csv('x_train.csv')
y_train = pd.read_csv('y_train.csv')
x_test = pd.read_csv('x_test.csv')
y_test = pd.read_csv('y_test.csv')

print(x_train.head(3))
print(f'shape x train: {x_train.shape}')
print(y_train.head(3))
print(f'shape y train: {y_train.shape}')
```

```python
## Membangun model sequential
model = Sequential()

# Menambahkan layer Dense dengan 64 unit, menggunakan fungsi aktivasi ReLU
model.add(Dense(64, activation='relu', input_shape=(5,)))

# Menambahkan layer Dense dengan 64 unit, menggunakan fungsi aktivasi ReLU
model.add(Dense(64, activation='relu'))

model.add(Dense(4, activation='softmax'))

# Mengompilasi model dengan optimizer Adam dan fungsi loss sparse_categorical_crossentropy
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Melatih model dengan data x_train dan y_train
model.fit(x_train, y_train, epochs=100, batch_size=32)

# Evaluasi model menggunakan data x_test dan y_test
loss, accuracy = model.evaluate(x_test, y_test)

# Menampilkan akurasi model
print("Accuracy:", accuracy)
```
